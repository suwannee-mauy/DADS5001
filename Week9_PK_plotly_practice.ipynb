```python
import pandas as pd
import plotly.express as px
import seaborn as sns

df = pd.read_csv('https://github.com/chris1610/pbpython/blob/master/data/cereal_data.csv?raw=True')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df['shelf'].unique()
```

```python
fig = px.scatter(df,
                x='sugars',
                y='rating',
                hover_name='name',
                #hover_name= df['name'].astype(str)+" "+df['mfr'].astype(str),  สามารถเปลี่ยนมาใช้ hover ได้
                title='Cereal_ratingsvs.sugars')
fig.show()
```

```python
df.dtypes
```

#ใช้สีตาม mfr และ size ตาม calories

```python
fig = px.scatter(df,
                x='sugars',
                y='rating',
                hover_name='name',
                #hover_name= df['name'].astype(str)+" "+df['mfr'].astype(str),  สามารถเปลี่ยนมาใช้ hover ได้
                 color="mfr",
                 size='calories',
                 title='Cereal_ratingsvs.sugars')
fig.show()
```

```python
fig = px.scatter(df,
                x='sugars',
                y='rating',
                hover_name='calories',
                #hover_name= df['name'].astype(str)+" "+df['mfr'].astype(str),  สามารถเปลี่ยนมาใช้ hover ได้
                 color="mfr",
                 size='fat',
                 title='Cereal_ratingsvs.sugars')
fig.show()
```

```python
fig = px.box(df,
                x='sugars',
                y='rating',
                hover_name='name',
                title='Cereal ratings vs. sugars')
fig.show()
```

```python
fig = px.histogram(df, x='rating', title='Rating distribution')#,nbins=7)
fig.show()
```

```python
fig = px.histogram(df, x='sugars', title='Rating distribution' ,nbins=2)  #ยังไม่โอเคน้าาา
fig.show()
```

```python
fig = px.scatter(df,
                x='sugars',
                y='rating',
                color='mfr',
                size='calories',
                facet_row='shelf', # แกน row ชั้นที่วาง
                facet_col='type',  # แกน Col
                hover_name='name',
                category_orders={'shelf': ['Top', 'Middle', 'Bottom']})
fig.show()
print(df.head())
```

```python
#concept เดียวกับ decision tree
fig = px.treemap(df, path=['mfr', 'shelf','calories'], values='cereal', title='Cereals by shelf location')   #title is information  path คือ การกำหนด layer เข้าไปข้างใน
fig.show()
```

```python
fig = px.treemap(df, path=['mfr', 'shelf','calories','potass'], values='cereal', title='Cereals by shelf location')   #title is information  path คือ การกำหนด layer เข้าไปข้างใน ลองเพิ่ม layer : 'potass'
fig.show()
```

```python
fig = px.sunburst(df, path=['mfr','calories','sugars'], values='cereal')
fig.show()
```
